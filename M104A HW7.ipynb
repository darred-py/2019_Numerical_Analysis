### Problem 2
source: https://docs.scipy.org/doc/numpy/reference/routines.fft.html#module-numpy.fft
Using numpy.fft.ifft:

Retuns values using $$f_j = \frac{1}{N}\sum^{N-1}_{k=0} c_k e^{2\pi ikj/N}$$

### Problem 3

Assume $f_j$ is real for $j = 0, 1, ..., N-1$. Then $$f_j = \Re[\frac{1}{N}\sum^{N-1}_{k=0} c_k e^{2\pi ikj/N}] = \frac{1}{N}\sum^{N-1}_{k=0} c_k e^{2\pi ikj/N}$$

### Problem 4

```python
import numpy as np
import matplotlib.pyplot as plt

# Code inlfuenced from 
# https://docs.scipy.org/doc/numpy/reference/generated/numpy.fft.ifft.html#numpy.fft.ifft

t = np.arange(400)

n = np.zeros((400,), dtype=complex)
m = np.zeros((400,), dtype=complex)

n[20:40] = np.exp(np.random.uniform(0, 2*np.pi, (20,)))
m[:20] = np.exp(np.random.uniform(0, 2*np.pi, (20,)))

s = np.fft.ifft(n)
u = np.fft.ifft(m)

plt.plot(t, s.real, 'b-', t, s.imag, 'r--')
plt.legend(('real', 'imaginary'))
plt.show()

plt.plot(t, u.real, 'b-', t, u.imag, 'r--')
plt.legend(('real', 'imaginary'))
plt.show()
```
