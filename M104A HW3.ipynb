### Problem 1
Let $u,v \in W$ such that $$M = \| f-u \| = \|f - v \|$$ where $M$ is the best approximation to $f$. Then 
$$M = \theta M + (1-\theta)M = \theta\|f-u\| + (1-\theta)\|f-v\|.$$ By the qualities of a norm, we can introduce $\theta$ into our norm since it is a constant: $$\|\theta(f-u)\| + \|(1-\theta)(f-v)\|.$$ Then, we proceed with triangle inequality:
$$\|\theta(f-u)\| + \|(1-\theta)(f-v)\| \geq \|\theta(f-u) + (1-\theta)(f-v)\|.$$ Thus, $$M \geq \|\theta(f-u) + (1-\theta)(f-v)\|,$$ which implies that $\theta(f-u) + (1-\theta)(f-v) \in W$.

### Problem 2

Then $$e_n(x) = f(x) = sin(2x).$$ We see that $sin(2x)$ has two local maxima at $x=\frac{\pi}{4}$ and at $x=\frac{5\pi}{4}$. Thus, $$\|e_2(\frac{\pi}{4})\|_\infty=\|e_2(\frac{5\pi}{4})\|_\infty$$ and $$\|e_2(\frac{3\pi}{4})\|_\infty=\|e_2(\frac{7\pi}{4})\|_\infty$$ where $$\|e_2(\frac{3\pi}{4})\|_\infty = -\|e_2(\frac{\pi}{4})\|_\infty.$$ Therefore, by Chebyshev's Equioscillation Theorem, $q=0$ is the best approxiamtion.

### Problem 3
Since $f(x)$ is continuous on a bounded interval, then it achieves it maximum and minimum at points $x_1$ and $x_2$, respectively. Then we choose as our constant $c$:
$$c = \frac{f(x_1) - f(x_2)}{2}.$$ Then our error of approximation becomes $$e_0(x) = f(x) - c$$.
Then, when $f(x)$ attains its maximum at $x_1$,
$$\|e_0(x_1)\|_\infty = \|f(x_1)-c\|_\infty = \frac{f(x_1) - f(x_2)}{2}.$$
Also, when $f(x)$ attains its minimum at $x_2$,
$$\|e_0(x_2)\|_\infty = \|f(x_2)-c\|_\infty = \frac{f(x_2) - f(x_1)}{2}.$$
Thus,
$$\|e_0(x_2)\| = -\|e_0(x_1)\|.$$
Therefore, since $e_0(x)$ equioscillates twice, then by Chebyshev's Equioscillation Theorem, we've attained the best approximation.

### Problem 5
Assume $p(x)$ has two Chebyshev Polynomial representations, $s(x)$ and $t(x)$. Then $$s(x) = p(x) = t(x) \implies s(x) - t(x) = 0$$
$$\iff \sum^n_{i=0}a_iT_i(x) - \sum^n_{i=0}b_iT_i(x) = \sum^n_{i=0}(a_i - b_i)T_i(x) = 0.$$ We define an inner product to be:
$$ \langle T_j(x), T_i(x) \rangle = \delta_{j,i}$$
where $\delta_{j,i} = 1$ for $j=i$ and $\delta_{j,i} = 0$ for $j \neq i$. Then by applying the inner product on either sides of our summations:
$$\sum^n_{i=0}(a_i - b_i) \langle T_i(x), T_n(x) \rangle = \langle 0,T_n(x) \rangle$$
$$\implies 0 = a_n - b_n \iff a_n = b_n.$$
Therefore, every polynomial $p(x)$ of degree $n$ has a unique Chebyshev Polynomial represenation.
